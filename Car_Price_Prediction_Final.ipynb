```python
def get_Brand(car_name):
    car_name= car_name.split(' ')[0] #need to explain
    return car_name.strip()
```

```python
def clean_data(value):
    value= value.split(' ')[0]
    value = value.strip()
    if value == ' ':
        value = 0
    try:
        # Try to convert to float
        return float(value)
    except ValueError:
        # If conversion fails, return 0 or NaN
        return 0.0
    return float(value)
```

```python
get_Brand('Maruti Swift Dzire VDI')
```

car_name.split(' ')[0]

car_name.split(' '): This splits the string car_name into a list of words, using a space ' ' as the delimiter. For example, if car_name = "Toyota Corolla", the result would be ['Toyota', 'Corolla'].
[0]: This selects the first word from the list. In our example, it would pick 'Toyota'. The first word is often the brand name of the car.
car_name.strip(' ')

This removes any leading or trailing spaces (whitespace) from the car_name string. The ' ' inside the strip() is actually unnecessary because strip() by default removes any kind of whitespace (spaces, tabs, newlines). For example, " Toyota ".strip() would return 'Toyota'.
